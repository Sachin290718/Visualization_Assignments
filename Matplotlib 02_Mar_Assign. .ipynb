# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

* Matplotlib is a powerful and versatile open-source plotting library for Python, designed to help users visualize data in a variety of formats.
* Pyplot is a module within Matplotlib that provides a MATLAB-like interface for making plots. It simplifies the process of adding plot elements such as lines, images, and text to the axes of the current figure.
* Key Uses of Matplotlib:

1) Basic Plots: Line plots, bar charts, histograms, scatter plots, etc.
2) Statistical Visualization: Box plots, error bars, and density plots.
3) Customization: Control over colors, labels, gridlines, and styles.
4) Subplots & Layouts: Create multiple plots in a single figure.
5) 3D Plotting: Surface plots and 3D scatter plots using mpl_toolkits.mplot3d.
6) Animations & Interactive Plots: Dynamic visualizations with FuncAnimation.
7) Integration: Works well with Pandas, NumPy and Jupyter Notebooks.

## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

* A scatter plot in Python is a type of data visualization that uses dots to represent values for two different numeric variables. It is used to observe and display the relationship between these variables, revealing patterns, trends, and correlations. Each dot's position on the horizontal (x-axis) and vertical (y-axis) axes corresponds to the values of the two variables for a single data point.
* Scatter plots are particularly useful in the initial stages of data analysis for exploring potential relationships between variables. They can help identify clusters, outliers, and trends that might not be apparent from raw data alone.

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

clr = np.random.rand(50)
plt.scatter(x,y, c= clr, s = 200, alpha = 0.5)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Sactter Plot')
plt.show()
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

* The subplot() function in Matplotlib is used to create multiple plots within a single figure, arranging them in a grid-like layout. This allows for efficient comparison and visualization of different datasets or variables side-by-side. Essentially, it divides a figure into smaller areas, where each area can hold its own plot. 

```python
# For line 1: 
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(2,2,1)
plt.plot(x1,y1, color='r', marker = 'o', markerfacecolor = 'b')

#For line 2: 
x1 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(2,2,2)
plt.plot(x1,y2, color = 'b', marker = '.', markerfacecolor = 'r')

#For line 3: 
x1 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(2,2,3)
plt.plot(x1,y3, color = 'c', marker = 's', markerfacecolor = 'm')

#For line 4: 
x1 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,4)
plt.plot(x1,y4, color = 'm' , marker = 'p' , markerfacecolor = 'c')

plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

* A bar plot, also known as a bar chart, is a graphical representation that uses rectangular bars to compare values across different categories.
* Why are bar plots used?
1) Comparing categories:
Bar plots are ideal for comparing values across different categories, such as sales figures for different products or customer demographics. 
2) Highlighting differences:
The visual length of the bars clearly shows the differences in values between categories, allowing for quick identification of key differences. 
3) Identifying trends:
Bar plots can be used to show changes in data over time or across different groups, revealing trends and patterns. 
4) Simplifying complex data:
Bar plots can help make complex data easier to understand by providing a concise visual representation of the relationships between categories and values. 
5) Categorical data visualization:
They are specifically designed to visualize categorical or nominal data, where the data is grouped into categories.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.xlabel("Name of Companies")
plt.ylabel("Profit")
plt.title("Bar Chart")
plt.show()
```

```python
plt.barh(company,profit)
plt.ylabel("Name of Companies")
plt.xlabel("Profit")
plt.title("Bar Chart")
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

* A box plot, also known as a box and whisker plot, is a visual representation of a dataset that summarizes key statistics, including the median, quartiles, and outliers. It's used to compare distributions between different groups and to understand the spread and potential skewness of the data.
*  Why it's used:
1) Comparing groups:
Box plots are excellent for visualizing and comparing the distributions of different groups or datasets, highlighting similarities and differences in their central tendencies, spread, and potential outliers. 
2) Identifying skewness:
The position of the median within the box and the lengths of the whiskers can provide insights into whether the data is skewed (unevenly distributed). 
3) Identifying outliers:
Box plots help visualize and identify data points that are significantly different from the rest of the data, which can be useful for understanding potential errors or unusual observations. 
4) Summarizing data:
Box plots provide a concise and visual summary of the data's key features, making it easier to grasp the distribution and identify potential patterns.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
x = [box1 , box2]
plt.boxplot(x, tick_labels = ['Box1','Box2'],patch_artist=True)
plt.show()
```

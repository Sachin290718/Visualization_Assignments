## Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

* Seaborn offers several versatile plots for visualizing data. Five commonly used ones include:
* Scatter plots (for visualizing relationships between two variables),
* Line plots (for showing trends over time),
* Histograms (for understanding the distribution of a single variable),
* Box plots (for comparing distributions of two or more groups), and
* Heatmaps (for visualizing data as a color-coded matrix).

# Uses of Plot
1. Scatter Plot:

* Identifying correlations between variables. 
* Exploring potential relationships, such as linear or non-linear relationships. 
* Detecting outliers.

2. Line Plot:

* Tracking changes in a variable over time. 
* Comparing different groups or categories of data. 
* Visualizing trends in sales, stock prices, or other time-series data.

3. Histogram:

* Understanding the shape of a distribution (e.g., normal, skewed, bimodal). 
* Identifying outliers or unusual values. 
* Comparing distributions across different groups.

4. Box Plot:

* Comparing distributions of a variable across different categories or groups.
* Identifying outliers and the range of data values. 
* Visualizing the spread and skewness of data.

5. Heatmap:

* Displaying correlation matrices, showing the relationships between multiple variables. 
* Visualizing data grids, such as heatmaps of stock prices or other financial data. 
* Creating visualizations of data with a large number of dimensions.

## Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
df = sns.load_dataset('fmri')
sns.lineplot(x= df['timepoint'], y= df['signal'], hue =df['event'] , style = df['region'])
```

```python
sns.lineplot(data = df, x= 'timepoint', y = 'signal', hue = 'region', style = 'event')
```

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
df= sns.load_dataset('titanic')

# 1st box plot

sns.boxplot(x='pclass', y= 'age', data=df)
plt.title("Age Distribution by Passanger Class")
plt.show()

# 2nd box plot

sns.boxenplot(x='pclass', y = 'fare' , data = df)
plt.title("Fare Distribution by Passanger Class")
plt.show()
```

## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
df = sns.load_dataset('diamonds')
sns.histplot(data=df, x='price', hue='cut',kde = True)
plt.show()
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
df = sns.load_dataset('iris')
sns.pairplot(data =df, hue = 'species')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

## Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
